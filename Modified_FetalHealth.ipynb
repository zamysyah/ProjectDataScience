4. Menentukan Objek Data

RINGKASAN
Objek Data:

baseline value: Nilai dasar denyut jantung janin (FHR) dalam denyut per menit (bpm).
accelerations: Jumlah akselerasi FHR per detik.
fetal_movement: Jumlah gerakan janin per detik.
uterine_contractions: Jumlah kontraksi uterus per detik.
light_decelerations: Jumlah deselerasi ringan FHR per detik.
severe_decelerations: Jumlah deselerasi berat FHR per detik.
prolongued_decelerations: Jumlah deselerasi berkepanjangan FHR per detik.
abnormal_short_term_variability: Persentase waktu dengan variabilitas jangka pendek yang abnormal.
mean_value_of_short_term_variability: Nilai rata-rata variabilitas jangka pendek.
percentage_of_time_with_abnormal_long_term_variability: Persentase waktu dengan variabilitas jangka panjang yang abnormal.
mean_value_of_long_term_variability: Nilai rata-rata variabilitas jangka panjang.
histogram_width: Lebar histogram FHR.
histogram_min: Nilai minimum histogram FHR.
histogram_max: Nilai maksimum histogram FHR.
histogram_number_of_peaks: Jumlah puncak dalam histogram FHR.
histogram_number_of_zero_crossings: Jumlah persilangan nol dalam histogram FHR.
histogram_mode: Modus histogram FHR.
histogram_mean: Rata-rata histogram FHR.
histogram_median: Median histogram FHR.
histogram_variance: Varians histogram FHR.
histogram_tendency: Kecenderungan histogram FHR (cth., datar, meningkat, menurun).
fetal_health: Klasifikasi kesehatan janin (1 = Normal, 2 = Suspect, 3 = Pathological).
Keterangan:

Objek data di atas merupakan fitur atau variabel yang diukur atau diamati dari data cardiotocography (CTG).
fetal_health adalah target variabel atau variabel dependen yang ingin diprediksi.
Fitur-fitur lainnya adalah variabel independen yang digunakan untuk memprediksi kesehatan janin.

```
data_iqr = dataset.copy() # Buat copy dataset asli
for col in dataset.select_dtypes(include=['float64']).columns:
    if col not in ['fetal_health']:  # Kecualikan kolom 'Fetal_healt' (karena bertipe kategorikal)
        Q1 = dataset[col].quantile(0.25)
        Q3 = dataset[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Ganti outliers dengan batas bawah/atas
        data_iqr[col] = dataset[col].apply(lambda x: lower_bound if x < lower_bound else (upper_bound if x > upper_bound else x))
        print(f"Outliers pada kolom '{col}' telah ditangani menggunakan metode IQR.")
```

```
# Pilih hanya kolom numerik selain kolom target
numerical_cols = data_iqr.select_dtypes(include=['float64']).columns
numerical_cols = numerical_cols[numerical_cols != 'fetal_health']  # Menghapus kolom 'fetal_health'

# Buat boxplot untuk setiap kolom numerik selain 'Fetal_healt'
plt.figure(figsize=(15, 10))
sns.boxplot(data=data_iqr[numerical_cols], orient='h', palette="Set2")
plt.title("Boxplot untuk Semua Kolom Numerik Selain 'fetal_health' (Setelah Handling Outliers)")
plt.show()
```

# 6.) Konstruksi Data

## Handling Imbalance Data

```
# Import Library yang dibutuhkan untuk handling imbalance
# Menggunakan metode oversampling
from imblearn.over_sampling import RandomOverSampler
from collections import Counter
from sklearn.model_selection import train_test_split
```

```
# Memisahkan kolom feature dan kolom target
X = data_iqr.drop("fetal_health",axis=1).values
y = data_iqr['fetal_health']
```

```
# Split data untuk training dan testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```
# Oversampling menggunakan RandomOverSampler
oversample = RandomOverSampler(sampling_strategy='auto', random_state=42)
X_over, y_over = oversample.fit_resample(X_train, y_train)
```

```
# Tampilkan distribusi kelas
"Distribusi kelas setelah oversampling:", Counter(y_over)
```

**VISUALISASI HANDLINIG**

```
# Visualisasi distribusi kelas sebelum handling imbalance
plt.figure(figsize=(8, 6))
sns.countplot(x=y_train, palette="Set2")
plt.title("Ditribusi kelas sebelum oversampling")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()

# Oversampling menggunakan RandomOverSampler
oversample = RandomOverSampler(sampling_strategy='auto', random_state=42)
X_over, y_over = oversample.fit_resample(X_train, y_train)

# Visualisasi distribusi kelas setelah oversampling
plt.figure(figsize=(8, 6))
sns.countplot(x=y_over, palette="Set2")
plt.title("Ditribusi kelas setelah oversampling")
plt.xlabel("fetal_health")
plt.ylabel("Count")
plt.show()
```

# 7.) Menentukan Label Data

```
#Persiapan untuk membangun model
#Menentukan Kolom independen dan dependen
ind_col = [col for col in dataset.columns if col != 'fetal_health']
dep_col = 'fetal_health'

#memisahkan fitur X dan target Y
x = dataset[ind_col]
y = dataset[dep_col]
```

```
# pembagian data ke dalam set pelatihan dan pengujian
from sklearn.model_selection import train_test_split

# Split data dengan perbandingan data training dan data testing adalah 70 : 30
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state= 0)
```

# 8.) Membangun dan Mengevaluasi Model

```
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```
# Inisialisasi
nb = GaussianNB()
rf = RandomForestClassifier(random_state=42)
svm = SVC(random_state=42)
lr = LogisticRegression(random_state=42)
dt = DecisionTreeClassifier(random_state=42)
knn = KNeighborsClassifier()

models = {
    "Naive Bayes": nb,
    "Random Forest": rf,
    "SVM": svm,
    "Logistic Regression": lr,
    "Decision Tree": dt,
    "K-Nearest Neighbors": knn,

}
```

### Naive Bayes

```
nb.fit(X_train, y_train)
```

```
nb_y_pred = nb.predict(X_test)
accuracy_nb = round(accuracy_score(y_test, nb_y_pred), 3)
```

```
print(confusion_matrix(y_test, nb_y_pred))
```

```
print(classification_report(y_test, nb_y_pred))
```

### Random Forest

```
rf.fit(X_train, y_train)
```

```
rf_y_pred = rf.predict(X_test)
accuracy_rf = round(accuracy_score(y_test, rf_y_pred), 3)
```

```
print(confusion_matrix(y_test, rf_y_pred))
```

```
print(classification_report(y_test, rf_y_pred))
```

### SVM

```
svm.fit(X_train, y_train)

svm_y_pred = svm.predict(X_test)
accuracy_svm = round(accuracy_score(y_test, svm_y_pred), 3)

print(confusion_matrix(y_test, svm_y_pred))

print(classification_report(y_test, svm_y_pred))
```

### Logistic Regression

```
lr.fit(X_train, y_train)

lr_y_pred = lr.predict(X_test)
accuracy_lr = round(accuracy_score(y_test, lr_y_pred), 3)

print(confusion_matrix(y_test, lr_y_pred))

print(classification_report(y_test, lr_y_pred))
```

### Decision Tree

```
dt.fit(X_train, y_train)

dt_y_pred = dt.predict(X_test)
accuracy_dt = round(accuracy_score(y_test, dt_y_pred), 3)

print(confusion_matrix(y_test, dt_y_pred))

print(classification_report(y_test, dt_y_pred))
```

### K-Nearest Neighbor

```
knn.fit(X_train, y_train)

knn_y_pred = knn.predict(X_test)
accuracy_knn = round(accuracy_score(y_test, knn_y_pred), 3)

print(confusion_matrix(y_test, knn_y_pred))

print(classification_report(y_test, knn_y_pred))

# Evaluasi skor akurasi, confusion matrix, dan classification report dari knn
knn_acc = accuracy_score(y_test, knn.predict(X_test))

print(f"Training Accuracy of KNN is {accuracy_score(y_train, knn.predict(X_train))}")
print(f"Test Accuracy of KNN is {knn_acc} \n")
```

```
models = ['Naive Bayes', 'Random Forest', 'SVM', 'Logistic Regression', 'Decission Tree', 'KNN']
accuracies = [
    accuracy_score(y_test, nb_y_pred),
    accuracy_score(y_test, rf_y_pred),
    accuracy_score(y_test, svm_y_pred),
    accuracy_score(y_test, lr_y_pred),
    accuracy_score(y_test, dt_y_pred),
    accuracy_score(y_test, knn_y_pred)
]
plt.figure(figsize=(10, 6))
sns.barplot(x=models, y=accuracies, palette='viridis')
for i, acc in enumerate(accuracies):
    plt.text(i, acc + 0.01, f'{acc:.2f}', ha='center', va='bottom', fontsize=10, color='black')
plt.title('Model Comparison - Accuracy', fontsize=16)
plt.ylabel('Accuracy', fontsize=12)
plt.xlabel('Models', fontsize=12)
plt.ylim(0, 1)
plt.show()
```
